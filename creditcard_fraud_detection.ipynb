### **Cleaning/Encoding**

```python
# Function to remove non-important columns
def clean_data(clean):
     clean.drop(["Unnamed: 0",'cc_num','first', 'last', 'street', 'city', 'state', 'zip', 'dob', 'trans_num','trans_date_trans_time'],axis=1, inplace=True)
     clean.dropna()
     return clean
```

```python
# Columns with categorical values
df_train.select_dtypes(include = ['object'])
```

```python
# Encoding to convert categorical data into numerical data
encoder=LabelEncoder()
def encode(data):
    data['merchant']=encoder.fit_transform(data['merchant'])
    data["category"] = encoder.fit_transform(data["category"])
    data["gender"] = encoder.fit_transform(data["gender"])
    data["job"] = encoder.fit_transform(data["job"])
    return data
```

```python
# Function called on training data
encode(df_train)
```

```python
# Function called on training data
encode(df_test)
```

```python
# Feature Engineering - Add new features that might help detect fraud
def add_features(df):
    # Calculate distance between customer and merchant
    df['distance'] = np.sqrt((df['lat'] - df['merch_lat'])**2 + (df['long'] - df['merch_long'])**2)

    # Time of day from unix timestamp (fraud may be more common at certain times)
    df['hour'] = pd.to_datetime(df['unix_time'], unit='s').dt.hour

    # Day of week
    df['day_of_week'] = pd.to_datetime(df['unix_time'], unit='s').dt.dayofweek

    # Amount bins (higher amounts may be more suspicious)
    df['amt_bin'] = pd.qcut(df['amt'], q=10, labels=False, duplicates='drop')

    return df
```

```python
df_train = add_features(df_train)
df_test = add_features(df_test)
```
